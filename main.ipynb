```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 1] = le.fit_transform(X[:, 1])    # labelencoding on sex variable 
```

```python
#X[:,1]
```

```python
df.sample(5)
```

```python
# onehot encoding over Embarked
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[6])],remainder='passthrough')
```

```python
X = np.array(ct.fit_transform(X))
```

```python
X
```

```python
# let's apply train test split 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#feature scaling to make the values in a range
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#X_train
```

```python
#X_test
```

```python
# let's build the ANN
ann = tf.keras.models.Sequential()
```

```python
# input and first hidden layer 
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))  
```

```python
# second hidden layer
```

```python
ann.add(tf.keras.layers.Dense(units=8,activation='relu'))
```

```python
# 3rd hidden layer with  10 neurons 
ann.add(tf.keras.layers.Dense(units=10,activation='relu'))
```

```python
# 4th hidden layer with 12 neurons 
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
```

```python
# output layer using sigmod cause we need to predict in binary
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))    
```

```python
# compiling the model
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy']) 
```

```python
# let's train the model 
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=200)
```

```python
# got the accuracy of 86%
```

```python
df.head()
```

```python
X_test
```

```python
ypred = ann.predict(X_test)
```

```python
ypred
```

```python
y_pred = (ypred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# let's make the  confusion matrix
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
acc = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy Score: {:.2f}".format(acc))
```
